### PROBLEM 1 (15 pts.; 50/50 for code & discussion) 
**Fibonacci sequence**

1. Write a function that returns the nth Fibonacci number. *(10 pts.)*

2. Print the nth Fibonacci number, for $n = 1, 10, 10^3, 10^6, 10^9$. *(5 pts.)*

**1.** I defined the function ``` fib ``` to take an input number $n$ and print the $nth$ term in the Fibinacci series. This utilizes a simple for loop to assign the new values for ``` a ``` and ``` b ```. As for loop runs, the new ``` a ``` value is the old ``` b ``` value, new ``` b ``` value is the sum of old values of ``` a ``` and ``` b ```. This continues till the end of range value ``` n ``` and the value of ``` b ``` is printed out

````
def fib(n):
  a,b=0,1
  for i in range (1,n):
    a,b=b,a+b
  return a
  print (b)
````

**2.** I called the function ``` fib ``` with arguments 1,10,1000,1000000, and 1000000000 to print the $1st,10th,10^3th,10^6th,$ and $10^9th$ Fibonacci term, respectively.

**Calculating the sum of a simple series**

In spite of the power of your trusty computer, calculating the sum of even a simple series may require some thought and care. Consider the two series

$$\begin{equation}
S^{\rm (up)} = \sum_{n=1}^{N} \frac{1}{n},\ S^{\rm (down)} = \sum_{n=N}^{1} \frac{1}{n}.
\end{equation}
$$

Both series are finite as long as $N$ is finite, and when summed analytically both give the same answer. Nonetheless, because of round-off error, the numerical value of $S^{\rm (up)}$ will not be precisely that of $S^{\rm (down)}$.

1. Calculate $S^{\rm (up)}$ and $S^{\rm (down)}$ as functions of $N$. *(10 pts.)*

2. Make a log-log plot of $(S^{\rm (up)}-S^{\rm (down)})/(|S^{\rm (up)}|+|S^{\rm (down)}|)$ vs. $N$. *(10 pts.)*

3. Observe the linear regime on your graph and explain why the downward sum is generally more precise. *(5 pts.)*

**1.** I define the function `summ()` to take an input of `N_final` for corresponding to $N$ and calculate the upward and downward sum

````
import numpy as np
import matplotlib.pyplot as plt


def summ(N_final):
  for N in range(1,N_final+1):
    up=0
    down=0
    for n in range(1,N+1):
      up+= 1/n
      down+= 1/(N+1-n)
    print("UP:", up, "DOWN", down)
````
